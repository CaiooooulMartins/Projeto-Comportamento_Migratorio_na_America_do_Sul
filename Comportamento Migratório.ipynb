Projeto de Comportamento Migratório na América do Sul

```python
import pandas as pd

df= pd.read_csv("imigrantes_canada.csv")
df  
df.info()
```

Analisar a tendencia de imigraçao do Brasil em um determinado periodo

```python
#Indexando a coluna País
df.set_index('País', inplace=True)
```

```python
#criando variavel para armazenar os anos para facilitar a manipulação dos dados

anos = list(map(str,range(1980,2014)))
anos
```

```python
#Filtrando o df para pegarmos apenas o Brasil

Brasil = df.loc ['Brasil',anos]
Brasil
```

```python
#Criando um data frame  a partir desse filtro que fizemos 

brasil_dict = {'ano': Brasil.index.tolist(), 'imigrantes': Brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
dados_brasil
```

```python
#Gerando o primeiro grafico 

import matplotlib.pyplot as plt 

plt.figure(figsize=(8,4))
plt.plot(dados_brasil["ano"],dados_brasil['imigrantes'] )
plt.title("Imigração do brasil para o Canada")
plt.xlabel("Ano")
plt.ylabel("Numero de Imigrantes")
plt.xticks(list(["1980","1985","1990","1995","2000","2005","2010"]))
plt.show

```

```python
#Cria uma figura
fig, ax = plt.subplots(figsize=(8,4))
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
```

```python
#Agora, vamos usar a função plt.show() para exibir somente a figura, sem as informações adicionais que o Python apresenta na plotagem.

fig, ax = plt.subplots(figsize=(8,4))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.show()
```

```python
#Agora separando os anos de 5 em 5 anos

fig, ax = plt.subplots(figsize=(8,4))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.show()
```

```python
#Adicionado titulo + rotulos - linhas 3 e 4

fig, ax = plt.subplots(figsize=(8,4))   
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'],lw=3,marker = 'o', color = 'g')    #lw - alterar expessura da linha // marker - serve para adicionar um marcador // color - define a cor
ax.set_title('Imigração do Brasil para o Canadá\n1980 a 2013',fontsize = 19,loc ='left')  #fontsize - serve para aumentar o tamanho da fonte // loc - a direão do texto
ax.set_xlabel('Ano',fontsize = 14)
ax.set_ylabel('Número de imigrantes',fontsize = 14)
ax.xaxis.set_tick_params(labelsize = 13)
ax.yaxis.set_tick_params(labelsize = 13)
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.grid(linestyle= '--')               #Adiciona uma grade para ajudar na vizualização
plt.show()
```

Analisar as tendências de imigração para os países da América do Sul, incluindo o Brasil. Será possível analisar mais de um país simultaneamente?

```python
fig,axs = plt.subplots(2,2, figsize = (10,6))
fig.subplots_adjust(hspace = 0.5, wspace = 0.3)  #AJUSTANDO ESPACAMENTO VERTICAL E HORIZONTAL 
fig.suptitle('Imigração dos quatro maiores paises da America do Sul para o Canadá de 1980 a 2013') #Adicionando um titulo ao conjunto de graficos 


axs[0,0].plot(df.loc['Brasil',anos]) #Sinalizando quais dados quero utilizar para gerar o grafico
axs[0,0].set_title('Brasil')#Adicionando um titulo ao graficos 

axs[0,1].plot(df.loc['Colômbia',anos])
axs[0,1].set_title('Colômbia')

axs[1,0].plot(df.loc['Argentina',anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Peru',anos])
axs[1,1].set_title('Peru')

for ax in axs.flat:                                         # Função 'For' para ajustar o eixo  X e separar de 5 em 5 anos e nao precisar imputar para cada grafico
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))



for ax in axs.flat:                                       # Função 'For' para adicionar os rotulos em todos os graficos e nao precisar imputar para cada grafico
    ax.set_xlabel('Ano')
    ax.set_ylabel('Número de imirantes')

#Ajustando as escalas do eixo Y para nao causar distorção na hora da vizualização dos graficos
ymin = 0           
ymax = 7000 

for ax in axs.ravel():
    ax.set_ylim(ymin,ymax)


plt.show()
```

Criando um grafico de barras para vizualizar a Imigração dos paises da America do sul para o Canadá 

```python
america_sul = df.query('Região == "América do Sul"')
america_sul

```

```python
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']


fig, ax = plt.subplots(figsize = (12,5))
ax.bar(america_sul.index,america_sul['Total'],color = cores )
ax.set_title('Imigração da America do Sul para o canada\n1980 a 2013',loc = 'left',fontsize =18)

ax.set_xlabel('')
ax.set_ylabel('Numero de imigrantes',fontsize = 14)
ax.xaxis.set_tick_params(labelsize =12 )
ax.yaxis.set_tick_params(labelsize =12 )
plt.show()
```

Criando grafico de barras invertido

```python
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']


fig, ax = plt.subplots(figsize = (12,5))
ax.barh(america_sul.index,america_sul['Total'],color = cores ) #o H apóps o bar faz com que o grafico seja invertido
ax.set_title('Imigração da America do Sul para o canada\n1980 a 2013',loc = 'left',fontsize =18)

ax.set_xlabel('Numero de imigrantes',fontsize = 14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize =12 )
ax.yaxis.set_tick_params(labelsize =12 )
plt.show()
```

```python
#ordenar os valores do maior para o menor

america_sul_sorted = america_sul.sort_values('Total',ascending = True)
```

```python
cores = []

for pais in america_sul_sorted.index: #funçao para destacar o Brasil no meio dos graficos
    if pais == 'Brasil':
        cores.append('green')
    else:
        cores.append('silver')


fig, ax = plt.subplots(figsize = (12,5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'],color = cores ) #o H apóps o bar faz com que o grafico seja invertido
ax.set_title('América do Sul: Brasil foi o quarto país com mais imigrantes\npara o Canadá no período de 1980 a 2013', loc='left', fontsize=16)

ax.set_xlabel('Numero de imigrantes',fontsize = 14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize =12 )
ax.yaxis.set_tick_params(labelsize =12 )


#Criando anotações no grafico, nesse caso colocando os valores de cada barra 
for i, v in enumerate(america_sul_sorted['Total']):
    ax.text(v + 20,i, str(v),color ='black',fontsize = 10, ha = 'left',va = 'center') # v+20 - Indica a posição onde o texto vai ser adicionado// v = valor correspondente a cada barra// i = posição y do texto// str(v) = converter o tamanho de cada barra em uma string, que vai ser o texto// ha = alinhamento horizontal do texto// va = define o alinhamento vertical do texto

#Retirando o frame  e deixando o grafico clean 
ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis= "both",which = "both",length = 0)

#salvando o grafico
fig.savefig('America_do_Sul_Brasil_quarto_Pais.png',transparent= False,dpi = 300, bbox_inches ='tight' )



plt.show()


```
